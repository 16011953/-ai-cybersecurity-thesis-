```python
#Qn 1
import numpy as np
import pandas as pd

# Simulate a dataset of cyber threats
np.random.seed(42)
num_threats = 1000
threat_types = ['phishing', 'malware', 'DDoS', 'insider', 'ransomware']
response_time = np.random.normal(loc=30, scale=10, size=num_threats)  # Response time in minutes

data = {
    'threat_type': np.random.choice(threat_types, size=num_threats),
    'detected': np.random.choice([True, False], size=num_threats, p=[0.85, 0.15]),
    'response_time': response_time
}

df = pd.DataFrame(data)

# Calculate effectiveness metrics
detection_rate = df['detected'].mean()
average_response_time = df[df['detected']]['response_time'].mean()

# Print results
print(f"Detection Rate: {detection_rate * 100:.2f}%")
print(f"Average Response Time: {average_response_time:.2f} minutes")

# Further analysis by threat type
effectiveness_by_threat_type = df.groupby('threat_type').agg({
    'detected': 'mean',
    'response_time': 'mean'
}).reset_index()

effectiveness_by_threat_type.columns = ['threat_type', 'detection_rate', 'average_response_time']

print("\nEffectiveness by Threat Type:")
print(effectiveness_by_threat_type)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate a dataset of cyber threats
np.random.seed(42)
num_threats = 1000
threat_types = ['phishing', 'malware', 'DDoS', 'insider', 'ransomware']
detection_methods = ['AI', 'Traditional']

data = {
    'threat_type': np.random.choice(threat_types, size=num_threats),
    'detection_method': np.random.choice(detection_methods, size=num_threats, p=[0.7, 0.3]),
    'detected': np.random.choice([True, False], size=num_threats, p=[0.85, 0.15]),
    'predicted': np.random.choice([True, False], size=num_threats, p=[0.80, 0.20]),
    'response_time': np.random.normal(loc=30, scale=10, size=num_threats)  # Response time in minutes
}

df = pd.DataFrame(data)

# Key strengths and weaknesses of existing AI models
strengths_weaknesses = {
    'Strengths': ['High Detection Rate', 'Quick Response Time'],
    'Weaknesses': ['False Positives', 'Cost']
}

# Performance of AI-based protection vs traditional methods
performance = df.groupby('detection_method').agg({
    'detected': 'mean',
    'response_time': 'mean'
}).reset_index()

# AI models prediction accuracy
prediction_accuracy = df['predicted'].mean()

# Enhancement of security by AI vs Traditional
ai_vs_traditional = df.groupby(['detection_method', 'detected']).size().unstack(fill_value=0)

# Plotting the results

# Pie chart for strengths and weaknesses
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Strengths
axs[0, 0].pie([1]*len(strengths_weaknesses['Strengths']), labels=strengths_weaknesses['Strengths'], autopct='%1.1f%%')
axs[0, 0].set_title('Strengths of AI Models')

# Weaknesses
axs[0, 1].pie([1]*len(strengths_weaknesses['Weaknesses']), labels=strengths_weaknesses['Weaknesses'], autopct='%1.1f%%')
axs[0, 1].set_title('Weaknesses of AI Models')

# Performance comparison
axs[1, 0].pie(performance['detected'], labels=performance['detection_method'], autopct='%1.1f%%')
axs[1, 0].set_title('Detection Performance of AI vs Traditional')

# AI vs Traditional enhancement
axs[1, 1].pie(ai_vs_traditional.sum(axis=1), labels=ai_vs_traditional.index, autopct='%1.1f%%')
axs[1, 1].set_title('AI vs Traditional Detection Rate')

plt.tight_layout()
plt.show()

# Print numerical results
print(f"Prediction Accuracy of AI Models: {prediction_accuracy * 100:.2f}%")
print("\nPerformance of AI vs Traditional Methods:")
print(performance)
print("\nAI vs Traditional Detection Count:")
print(ai_vs_traditional)
```

```python
import numpy as np
import pandas as pd

# Simulate a dataset of cyber threats
np.random.seed(42)
num_threats = 1000
threat_types = ['phishing', 'malware', 'DDoS', 'insider', 'ransomware']
detection_methods = ['AI', 'Traditional']

data = {
    'threat_id': np.arange(1, num_threats + 1),
    'threat_type': np.random.choice(threat_types, size=num_threats),
    'detection_method': np.random.choice(detection_methods, size=num_threats, p=[0.7, 0.3]),
    'detected': np.random.choice([True, False], size=num_threats, p=[0.85, 0.15]),
    'predicted': np.random.choice([True, False], size=num_threats, p=[0.80, 0.20]),
    'response_time': np.random.normal(loc=30, scale=10, size=num_threats).clip(min=5),
    'mitigation_success': np.random.choice([True, False], size=num_threats, p=[0.90, 0.10]),
    'false_positive': np.random.choice([True, False], size=num_threats, p=[0.10, 0.90]),
    'cost': np.random.normal(loc=10000, scale=5000, size=num_threats).clip(min=1000)
}

df = pd.DataFrame(data)

# Print sample data
print(df.head())

# Calculate effectiveness metrics
detection_rate = df['detected'].mean()
average_response_time = df[df['detected']]['response_time'].mean()
prediction_accuracy = df['predicted'].mean()
mitigation_success_rate = df['mitigation_success'].mean()
false_positive_rate = df['false_positive'].mean()
average_cost = df['cost'].mean()

# Group data by detection method
performance = df.groupby('detection_method').agg({
    'detected': 'mean',
    'response_time': 'mean',
    'mitigation_success': 'mean',
    'false_positive': 'mean',
    'cost': 'mean'
}).reset_index()

# AI vs Traditional detection count
ai_vs_traditional = df.groupby(['detection_method', 'detected']).size().unstack(fill_value=0)

# Print numerical results
print(f"Overall Detection Rate: {detection_rate * 100:.2f}%")
print(f"Average Response Time: {average_response_time:.2f} minutes")
print(f"Prediction Accuracy of AI Models: {prediction_accuracy * 100:.2f}%")
print(f"Mitigation Success Rate: {mitigation_success_rate * 100:.2f}%")
print(f"False Positive Rate: {false_positive_rate * 100:.2f}%")
print(f"Average Cost: ${average_cost:.2f}")

print("\nPerformance of AI vs Traditional Methods:")
print(performance)
print("\nAI vs Traditional Detection Count:")
print(ai_vs_traditional)

# Visualizations
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Pie chart for detection methods
axs[0, 0].pie([detection_rate, 1 - detection_rate], labels=['Detected', 'Not Detected'], autopct='%1.1f%%')
axs[0, 0].set_title('Overall Detection Rate')

# Pie chart for prediction accuracy
axs[0, 1].pie([prediction_accuracy, 1 - prediction_accuracy], labels=['Predicted', 'Not Predicted'], autopct='%1.1f%%')
axs[0, 1].set_title('Prediction Accuracy of AI Models')

# Bar graph for detection performance comparison
axs[1, 0].bar(performance['detection_method'], performance['detected'] * 100, color=['blue', 'orange'])
axs[1, 0].set_title('Detection Rate by Method')
axs[1, 0].set_ylabel('Detection Rate (%)')

# Bar graph for response time comparison
axs[1, 1].bar(performance['detection_method'], performance['response_time'], color=['blue', 'orange'])
axs[1, 1].set_title('Average Response Time by Method')
axs[1, 1].set_ylabel('Response Time (minutes)')

# Bar graph for mitigation success comparison
axs[2, 0].bar(performance['detection_method'], performance['mitigation_success'] * 100, color=['blue', 'orange'])
axs[2, 0].set_title('Mitigation Success Rate by Method')
axs[2, 0].set_ylabel('Mitigation Success Rate (%)')

# Bar graph for cost comparison
axs[2, 1].bar(performance['detection_method'], performance['cost'], color=['blue', 'orange'])
axs[2, 1].set_title('Average Cost by Method')
axs[2, 1].set_ylabel('Average Cost ($)')

plt.tight_layout()
plt.show()
```
